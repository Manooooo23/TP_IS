<h3><center>EXERCICE 1 : ÉCHANTILLONNAGE</center></h3>

On considère le signal analogique $y(t) = cos(2πf0t)$ avec $t$ $∈$ $R$, et $f0 = 3kHz$. On échantillonne $y(t)$ à la fréquence $fe$.
Soit $ye(n)$ le signal ainsi échantillonné.

1) On fixe $fe = 10kHz$. Représenter le signal temporel et le module de sa transformée de Fourier sur deux figures différentes.

```python
f0 = 3000
fe = 10000
nb_samples = 100

samples = np.arange(nb_samples)
ye = np.cos(2 * np.pi * f0 / fe * samples)

# Signal temporel
plt.figure(figsize=(10, 4))
plt.stem(samples, ye, markerfmt='C0o', linefmt='C0-', basefmt='k-')
plt.title(f'Signal échantillonné ye(n), fe = {fe} Hz, f0 = {f0} Hz')
plt.xlabel('n (échantillons)')
plt.ylabel('ye(n)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Module de la transformée de Fourier
Ye = np.fft.fft(ye)
freqs = np.fft.fftfreq(nb_samples, d=1/fe)

plt.figure(figsize=(10, 4))
plt.plot(np.fft.fftshift(freqs), np.abs(np.fft.fftshift(Ye)))
plt.title(f'Module de la TF de ye(n), fe = {fe} Hz, f0 = {f0} Hz')
plt.xlabel('Fréquence (Hz)')
plt.ylabel('|Ye(f)|')
plt.grid(True)
plt.tight_layout()
plt.show()
```

2) On fixe $fe = 4kHz$. Représenter le signal temporel et le module de sa transformée de Fourier sur deux figures différentes. Conclure.

```python
f0 = 3000
fe = 4000
nb_samples = 100

samples = np.arange(nb_samples)
ye = np.cos(2 * np.pi * f0 / fe * samples)

# Signal temporel
plt.figure(figsize=(10, 4))
plt.stem(samples, ye, markerfmt='C0o', linefmt='C0-', basefmt='k-')
plt.title(f'Signal échantillonné ye(n), fe = {fe} Hz, f0 = {f0} Hz')
plt.xlabel('n (échantillons)')
plt.ylabel('ye(n)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Module de la transformée de Fourier
Ye = np.fft.fft(ye)
freqs = np.fft.fftfreq(nb_samples, d=1/fe)

plt.figure(figsize=(10, 4))
plt.plot(np.fft.fftshift(freqs), np.abs(np.fft.fftshift(Ye)))
plt.title(f'Module de la TF de ye(n), fe = {fe} Hz, f0 = {f0} Hz')
plt.xlabel('Fréquence (Hz)')
plt.ylabel('|Ye(f)|')
plt.grid(True)
plt.tight_layout()
plt.show()
```

*Commentaires (fréquence d'échantillonnage)*: 

Avec $f_e = 4$ kHz et $f_0 = 3$ kHz, le théorème de Shannon-Nyquist n'est pas respecté car $f_e < 2f_0$ (4000 < 6000 Hz). Il y a donc **repliement spectral (aliasing)** : le signal est perçu comme un cosinus à la fréquence $f_e - f_0 = 1$ kHz au lieu de 3 kHz. Le spectre montre des pics à ±1000 Hz et non ±3000 Hz.

3) Charger, à l’aide de scipy.io.wavfile, le son "handel" qui a été échantillonné à la fréquence d’échantillonnage $fe = 8192Hz$ et quantifié sur 16 bits.

```python
from scipy.io import wavfile
from IPython.display import Audio, display

# Chargement du fichier handel.wav
fe, handel = wavfile.read('handel.wav')

print(f"Fréquence d'échantillonnage : {fe} Hz")
print(f"Nombre d'échantillons : {len(handel)}")
print(f"Durée : {len(handel)/fe:.2f} s")
print(f"Type de données : {handel.dtype}")

# Écouter le signal original
display(Audio(handel, rate=fe))
```

4) Sous-échantillonner le signal sonore précédent d’un facteur 2 et écoutez-le. Conclure sur les effets du sous-échantionllonnage.

```python
# Sous-échantillonnage par un facteur 2 : on garde 1 échantillon sur 2
facteur = 2
handel_sous = handel[::facteur]
fe_sous = fe // facteur  # nouvelle fréquence d'échantillonnage

print(f"Fréquence originale      : {fe} Hz  →  Nombre d'échantillons : {len(handel)}")
print(f"Fréquence sous-échant.   : {fe_sous} Hz  →  Nombre d'échantillons : {len(handel_sous)}")

# --- Signal temporel ---
t_orig = np.arange(len(handel)) / fe
t_sous = np.arange(len(handel_sous)) / fe_sous

fig, axes = plt.subplots(2, 1, figsize=(12, 6), sharex=True)
axes[0].plot(t_orig[:1000], handel[:1000], label=f'Original  (fe={fe} Hz)', color='steelblue')
axes[0].set_title('Signal original (premiers 1000 échantillons)')
axes[0].set_ylabel('Amplitude')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(t_sous[:500], handel_sous[:500], label=f'Sous-échantillonné (fe={fe_sous} Hz)', color='tomato')
axes[1].set_title('Signal sous-échantillonné (premiers 500 échantillons)')
axes[1].set_xlabel('Temps (s)')
axes[1].set_ylabel('Amplitude')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

# --- Spectres ---
N_orig = len(handel)
N_sous = len(handel_sous)

HANDEL_fft  = np.fft.fftshift(np.abs(np.fft.fft(handel.astype(float))))
SOUS_fft    = np.fft.fftshift(np.abs(np.fft.fft(handel_sous.astype(float))))

freqs_orig = np.fft.fftshift(np.fft.fftfreq(N_orig, d=1/fe))
freqs_sous = np.fft.fftshift(np.fft.fftfreq(N_sous, d=1/fe_sous))

fig, axes = plt.subplots(2, 1, figsize=(12, 6))
axes[0].plot(freqs_orig, HANDEL_fft, color='steelblue')
axes[0].set_title(f'Spectre du signal original (fe = {fe} Hz)')
axes[0].set_xlabel('Fréquence (Hz)')
axes[0].set_ylabel('|X(f)|')
axes[0].grid(True)

axes[1].plot(freqs_sous, SOUS_fft, color='tomato')
axes[1].set_title(f'Spectre du signal sous-échantillonné (fe = {fe_sous} Hz)')
axes[1].set_xlabel('Fréquence (Hz)')
axes[1].set_ylabel('|X(f)|')
axes[1].grid(True)

plt.tight_layout()
plt.show()

# --- Écoute ---
print("Signal sous-échantillonné :")
display(Audio(handel_sous, rate=fe_sous))
```

*Commentaires (effets du sous-échantionnage)*:

Le sous-échantillonnage par un facteur 2 ramène la fréquence d'échantillonnage de $f_e = 8192$ Hz à $f_e' = 4096$ Hz. La fréquence de Nyquist passe alors de $4096$ Hz à $2048$ Hz. Or le signal "handel" contient des composantes fréquentielles allant jusqu'à $4096$ Hz : toutes les fréquences comprises entre $2048$ Hz et $4096$ Hz se retrouvent **repliées** (aliasing) dans le spectre $[0, 2048]$ Hz. À l'écoute, le signal semble **plus grave et plus "métallique"** : les hautes fréquences sont déformées par le repliement spectral, ce qui altère la qualité sonore.

5) Sur-échantillonner le signal sonore précédent d’un facteur 2 et écoutez-le. Conclure sur les effets du sur-échantionllonnage.

```python
# Sur-échantillonnage par interpolation linéaire (facteur 2)
facteur = 2
n_orig = np.arange(len(handel))
n_sur = np.linspace(0, len(handel) - 1, len(handel) * facteur)
handel_sur = np.interp(n_sur, n_orig, handel.astype(float))
fe_sur = fe * facteur

print(f"Fréquence originale      : {fe} Hz  →  Nombre d'échantillons : {len(handel)}")
print(f"Fréquence sur-échant.    : {fe_sur} Hz  →  Nombre d'échantillons : {len(handel_sur)}")

# --- Signal temporel ---
t_orig = np.arange(len(handel)) / fe
t_sur = np.arange(len(handel_sur)) / fe_sur

fig, axes = plt.subplots(2, 1, figsize=(12, 6), sharex=True)
axes[0].plot(t_orig[:1000], handel[:1000], label=f'Original (fe={fe} Hz)', color='steelblue')
axes[0].set_title('Signal original (premiers 1000 échantillons)')
axes[0].set_ylabel('Amplitude')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(t_sur[:2000], handel_sur[:2000], label=f'Sur-échantillonné (fe={fe_sur} Hz)', color='forestgreen')
axes[1].set_title('Signal sur-échantillonné (premiers 2000 échantillons)')
axes[1].set_xlabel('Temps (s)')
axes[1].set_ylabel('Amplitude')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

# --- Spectres ---
N_orig = len(handel)
N_sur = len(handel_sur)

HANDEL_fft = np.fft.fftshift(np.abs(np.fft.fft(handel.astype(float))))
SUR_fft    = np.fft.fftshift(np.abs(np.fft.fft(handel_sur)))

freqs_orig = np.fft.fftshift(np.fft.fftfreq(N_orig, d=1/fe))
freqs_sur  = np.fft.fftshift(np.fft.fftfreq(N_sur, d=1/fe_sur))

fig, axes = plt.subplots(2, 1, figsize=(12, 6))
axes[0].plot(freqs_orig, HANDEL_fft, color='steelblue')
axes[0].set_title(f'Spectre du signal original (fe = {fe} Hz)')
axes[0].set_xlabel('Fréquence (Hz)')
axes[0].set_ylabel('|X(f)|')
axes[0].grid(True)

axes[1].plot(freqs_sur, SUR_fft, color='forestgreen')
axes[1].set_title(f'Spectre du signal sur-échantillonné (fe = {fe_sur} Hz)')
axes[1].set_xlabel('Fréquence (Hz)')
axes[1].set_ylabel('|X(f)|')
axes[1].grid(True)

plt.tight_layout()
plt.show()

# --- Écoute ---
print("Signal sur-échantillonné :")
display(Audio(handel_sur.astype(np.int16), rate=fe_sur))
```

*Commentaires (effets du sur-échantionllonnage)*: 

Le sur-échantillonnage par un facteur 2 double la fréquence d'échantillonnage ($f_e' = 16384$ Hz) et le nombre d'échantillons, mais **n'ajoute aucune information fréquentielle** nouvelle. Le spectre du signal conserve le même contenu fréquentiel, simplement étalé sur une bande plus large $[-f_e'/2, f_e'/2]$. À l'écoute, le son est **identique** à l'original : le sur-échantillonnage ne modifie pas la perception du signal, contrairement au sous-échantillonnage qui introduit de l'aliasing.

<h3><center>EXERCICE 2 : TRANSFORMÉE DE FOURIER ET FILTRAGE</center></h3>

Soit le signal $x(n) = (1/2)^n$  $u(n)$, avec $u(n)$ l’échelon unité. 

1) Calculer $X(f)$ la transformée de Fourier de $x(n)$, avec $n = −128, ..., −1, 0, 1, 2, ..., 128$. Représenter le module et la phase de $X(f)$. On rappelle que $f$ $∈$ $[$−1/2$,  $1/2$]$



```python
n = np.arange(-128, 129)
u = (n >= 0).astype(float)
x = (0.5 ** n) * u

N = len(n)
X = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(x)))
f = np.fft.fftshift(np.fft.fftfreq(N))

fig, axes = plt.subplots(2, 1, figsize=(10, 6))

axes[0].plot(f, np.abs(X))
axes[0].set_title('Module de X(f)')
axes[0].set_xlabel('Fréquence normalisée f')
axes[0].set_ylabel('|X(f)|')
axes[0].grid(True)

axes[1].plot(f, np.angle(X))
axes[1].set_title('Phase de X(f)')
axes[1].set_xlabel('Fréquence normalisée f')
axes[1].set_ylabel('∠X(f) (rad)')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

2) Quelle est la nature (passe-haut, bas, bande) de ce filtre?

*Commentaires*:

Le module de $X(f)$ est maximal en $f = 0$ et décroît lorsque $|f|$ augmente. Il s'agit donc d'un **filtre passe-bas**.

Soit $y(n) = cos(2πf1n)+cos(2πf2n)+cos(2πf3n)$, avec $n = −128, ..., −1, 0, 1, 2, ..., 128$ et $f1 = 1/5$, $f2 = 1/3$, $f3 = 2/5$.

4) Représenter y(n).

```python
n = np.arange(-128, 129)
f1, f2, f3 = 1/5, 1/3, 2/5
y = np.cos(2*np.pi*f1*n) + np.cos(2*np.pi*f2*n) + np.cos(2*np.pi*f3*n)

plt.figure(figsize=(12, 4))
plt.plot(n, y)
plt.title(r'$y(n) = \cos(2\pi f_1 n) + \cos(2\pi f_2 n) + \cos(2\pi f_3 n)$')
plt.xlabel('n')
plt.ylabel('y(n)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

5) Quelle est la fréquence d’échantillonnage à laquelle est échantillonné le signal $y(n)$ ?

*Commentaires*:

Le signal $y(n)$ est un signal en temps discret. Les fréquences $f_1, f_2, f_3$ sont des **fréquences normalisées** (en cycles/échantillon). La fréquence d'échantillonnage est donc $f_e = 1$ Hz (normalisée), et la fréquence de Nyquist est $f_e/2 = 1/2$.

6) On souhaite éliminer les fréquences $f2$ et $f3$.
Donner la nature et la fréquence de coupure du filtre à utiliser.

*Commentaires*:

Pour éliminer $f_2 = 1/3$ et $f_3 = 2/5$ tout en conservant $f_1 = 1/5$, on utilise un **filtre passe-bas** avec une fréquence de coupure $f_c$ telle que $f_1 < f_c < f_2$, soit par exemple $f_c = 1/4 = 0.25$.

7) Réaliser le filtrage. Afficher le signal filtré temporel et sa transformée de Fourier sur deux figures différentes. Conclure.

```python
n = np.arange(-128, 129)
f1, f2, f3 = 1/5, 1/3, 2/5
y = np.cos(2*np.pi*f1*n) + np.cos(2*np.pi*f2*n) + np.cos(2*np.pi*f3*n)

N = len(n)
Y = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(y)))
f = np.fft.fftshift(np.fft.fftfreq(N))

# Filtre passe-bas idéal fc = 1/4
H_lp = (np.abs(f) <= 0.25).astype(float)

# Filtrage dans le domaine fréquentiel
Y_filtre = Y * H_lp
y_filtre = np.real(np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(Y_filtre))))

# Signal filtré
plt.figure(figsize=(12, 4))
plt.plot(n, y_filtre)
plt.title('Signal filtré (passe-bas, fc = 1/4)')
plt.xlabel('n')
plt.ylabel('y_filtré(n)')
plt.grid(True)
plt.tight_layout()
plt.show()

# TF du signal filtré
plt.figure(figsize=(12, 4))
plt.plot(f, np.abs(Y_filtre))
plt.title('Module de la TF du signal filtré')
plt.xlabel('Fréquence normalisée f')
plt.ylabel('|Y_filtré(f)|')
plt.grid(True)
plt.tight_layout()
plt.show()
```

*Commentaires*: 

Le filtre passe-bas à $f_c = 1/4$ a bien éliminé les composantes à $f_2 = 1/3$ et $f_3 = 2/5$. Le signal filtré ne contient plus que la composante à $f_1 = 1/5$ : on observe un cosinus pur dans le domaine temporel, et deux pics à $\pm 1/5$ dans le spectre.

8) Comment modifier le filtre passe-bas idéal pour éliminer $f1$ ?

*Commentaires*: 

Pour éliminer $f_1 = 1/5$ tout en conservant $f_2$ et $f_3$, il suffit de transformer le filtre passe-bas en **filtre passe-haut** : $H_{ph}(f) = 1 - H_{pb}(f)$. Avec la même fréquence de coupure $f_c = 1/4$, ce filtre laisse passer les fréquences $|f| > 1/4$ et bloque celles en dessous.

9) Réaliser le filtrage. Afficher le signal filtré temporel et sa transformée de Fourier sur deux figures différentes. Conclure.

```python
n = np.arange(-128, 129)
f1, f2, f3 = 1/5, 1/3, 2/5
y = np.cos(2*np.pi*f1*n) + np.cos(2*np.pi*f2*n) + np.cos(2*np.pi*f3*n)

N = len(n)
Y = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(y)))
f = np.fft.fftshift(np.fft.fftfreq(N))

# Filtre passe-haut idéal fc = 1/4
H_hp = 1 - (np.abs(f) <= 0.25).astype(float)

# Filtrage
Y_filtre_hp = Y * H_hp
y_filtre_hp = np.real(np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(Y_filtre_hp))))

# Signal filtré
plt.figure(figsize=(12, 4))
plt.plot(n, y_filtre_hp)
plt.title('Signal filtré (passe-haut, fc = 1/4)')
plt.xlabel('n')
plt.ylabel('y_filtré(n)')
plt.grid(True)
plt.tight_layout()
plt.show()

# TF du signal filtré
plt.figure(figsize=(12, 4))
plt.plot(f, np.abs(Y_filtre_hp))
plt.title('Module de la TF du signal filtré (passe-haut)')
plt.xlabel('Fréquence normalisée f')
plt.ylabel('|Y_filtré(f)|')
plt.grid(True)
plt.tight_layout()
plt.show()
```

*Commentaires*: 

Le filtre passe-haut à $f_c = 1/4$ a éliminé la composante $f_1 = 1/5$ et conservé $f_2 = 1/3$ et $f_3 = 2/5$. Le spectre montre des pics uniquement à $\pm 1/3$ et $\pm 2/5$, confirmant le bon fonctionnement du filtrage.
